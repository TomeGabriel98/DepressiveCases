```python
import pandas as pd
import matplotlib.pyplot as plt


dados = pd.read_csv("./CasosDepressivosV2.csv", low_memory=False,delimiter=';', encoding='UTF-8')

dados.head(39)
```

```python
dadosMascPorIdade = dados.groupby(by=['IDADE'])['TOTAL_MASC'].sum()
dadosFemPorIdade = dados.groupby(by=['IDADE'])['TOTAL_FEM'].sum()
dadosPorIdade = dados.groupby(by=['IDADE'])['TOTAL_CASOS'].sum()
dadosPorRegiao = dados.groupby(by=['REGIAO'])['TOTAL_CASOS'].sum()

print(dadosMascPorIdade, dadosFemPorIdade, dadosPorIdade, dadosPorRegiao)
```

```python
import matplotlib.pyplot as plt
homens = dadosMascPorIdade.values
mulheres = dadosFemPorIdade.values
total = dadosPorIdade.values
plt.plot(total,  label = 'Total')
plt.plot(mulheres,  label = 'Mulheres')
plt.plot(homens, label = 'Homens')

idades = dadosPorIdade.keys()

plt.title("Casos Depressivos - 2016 (Por idade)")
plt.xlabel("Idade")
plt.ylabel("Quantidade")
plt.xticks(range(8), (idades) , rotation=30 )
plt.grid()
plt.legend()
plt.show()
```

```python
plt.title("Casos Depressivos - 2016 (Por Regiao)")
plt.xlabel("Região")
plt.ylabel("Quantidade")
plt.bar(range(len(dadosPorRegiao)), dadosPorRegiao.values, width=0.8, color='blue', align='center')
plt.xticks(range(len(dadosPorRegiao)), (dadosPorRegiao.keys()) , rotation=30)
plt.grid()
plt.show()
```
